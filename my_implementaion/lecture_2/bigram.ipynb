```python
import torch
torch.set_printoptions(sci_mode=False)
```

```python
with open('names.txt','r') as f:
    x = f.read()

next_char=[]
for i in x.split('\n'):
    i = '.'+i+'.'
    for j in range(len(i)-1):
        next_char.append(i[j]+i[j+1])
        
```

```python
count_next_char = {}
for i in next_char:
    if i in count_next_char:
        count_next_char[i] += 1
    else:
        count_next_char[i] = 1
```

```python
alphaToNum = {'.':0}
numToAlpha = {0:'.'}
count = 1
for char in range(ord('a'), ord('z') + 1):
    # print(chr(char))
    numToAlpha[count] = chr(char)
    alphaToNum[chr(char)] = count
    count += 1
```

```python
t = torch.tensor([])
ones = t.new_ones((27,27))
for i in next_char:
    ones[alphaToNum[i[0]]][alphaToNum[i[1]]] += 1.0

prob = ones.clone()
for i in range(27):

    s = prob[i].sum()
    for j in range(27):
        prob[i][j] = prob[i][j]/ s

    
    
```

```python
for i in range(30):
    start=0
    name = ''
    while True:
        start = torch.multinomial(prob[start],1,True).item()
        if not start:
            break
        name += numToAlpha[start]
    print(name)
```
